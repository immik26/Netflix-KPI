```python
import pandas as pd
```

```python
df = pd.read_csv(r"C:\Users\irfan\netflix1.csv\netflix1.csv")
df.head()
```

```python
df.info()
df.describe()
df.isnull().sum()
```

```python
df['director'].fillna("Unknown", inplace=True)
df['country'].fillna("Unknown", inplace=True)
df['rating'].fillna("Not Rated", inplace=True)
```

```python
df.dropna(subset=['date_added'], inplace=True)
```

```python
df.drop_duplicates(inplace=True)
```

```python
df['date_added'] = pd.to_datetime(df['date_added'])
```

```python
df['year_added'] = df['date_added'].dt.year
df['month_added'] = df['date_added'].dt.month
```

```python
df['country'] = df['country'].str.strip()
df['genre'] = df['listed_in'].str.strip()
```

```python
df = df.assign(genre=df['listed_in'].str.split(',')).explode('genre')
```

```python
df['type'] = df['type'].str.title()
```

```python
df['duration_num'] = df['duration'].str.extract('(\d+)')
df['duration_num'] = pd.to_numeric(df['duration_num'])
```

```python
df.columns = df.columns.str.lower().str.replace(" ", "_")
```

```python
df.isnull().sum()
df.head()
```

```python
df.to_csv("netflix_cleaned.csv", index=False)
```
